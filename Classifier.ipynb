```python
import pandas as pd
```

```python
CPC2CPP_ACC = pd.read_csv('Classifier/CP_C2CP_P/ACC.csv', header=None)
CPC2CPP_LOSS = pd.read_csv('Classifier/CP_C2CP_P/loss.csv', header=None)

CPC2CPP_ACC = CPC2CPP_ACC.T
CPC2CPP_LOSS = CPC2CPP_LOSS.T

CPC2CPP_ACC = CPC2CPP_ACC.rename(columns={0 : 'ACC'})
CPC2CPP_LOSS = CPC2CPP_LOSS.rename(columns={0 : 'Loss'})

all = pd.concat([CPC2CPP_ACC, CPC2CPP_LOSS], axis=1)
```

```python
%matplotlib inline
all.plot(figsize=(15, 7)).line()
```
